# Reading data in pandas

```python
import pandas as pd
from io import StringIO
```

```python
emp_json = '{"id":1,"name":"Alice","department":"Sales","salary":50000,"join_date":"2019-08-15","location":{"city":"New York","country":"USA"}}'
```

```python
# reading from JSON

df = pd.read_json(StringIO(emp_json))
df
```

```python
# converting it back to json

df.to_json(orient='index')
```

```python
# reading csv from url

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', header=None)
df.head()
```

```python
# from html

url = "https://fdic.gov/bank-failures/failed-bank-list"
df = pd.read_html(url)
df[0]
```

```python
url = "https://en.wikipedia.org/wiki/Mobile_country_code"
df = pd.read_html(url, match="Country")
df[0]
```
